```python
review_count_df = disney_df.groupby(["UniqueID"]).size().reset_index()
review_count_df
```

```python
plt.figure(figsize=(10,6))
plt.bar(review_count_df['UniqueID'], review_count_df[0])
plt.xlabel('Movie Title')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews for each Movie')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_sentiment_df = disney_df.groupby(["UniqueID"])["compound"].mean().reset_index()
avg_sentiment_df
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_sentiment_df['UniqueID'], avg_sentiment_df["compound"])
plt.xlabel('Movie Title')
plt.ylabel('Average Sentiment')
plt.title('Mean Sentiment per Movie (-1 to 1 scale)')
plt.xticks(rotation=90)
plt.show()
```
