```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
disney_df = pd.read_csv("/content/Project1_DS4002/Data/FinalDisneyDataset.csv")
```

```python
disney_df
```

```python
disney_df['release_date'].unique()
```

```python
disney_df = disney_df.replace('Null', pd.NA).dropna()
disney_df['rating'] = disney_df['rating'].astype(float)
sns.barplot(x='UniqueID', y='rating', hue='recent?', data=disney_df, errorbar=None)
plt.xticks(rotation=90)
plt.title('Average Rating for Recent and Non-Recent Movies')
plt.legend(title='Recent?', loc='upper right')

```
